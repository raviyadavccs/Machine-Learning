### Encoding parameter

#### there are many dataset in which bydefualt encodeing is utf8 but there are some dataset which is encodeed in different format 


#### lets create a df with latin1 encoding


```python
df=pd.DataFrame({
    "name":["José", "André", "François", "Müller"],
    "city":["São Paulo", "Belo Horizonte", "Paris", "Berlin"]
})

df.to_csv("employee_latin1.csv", index=False,encoding="latin1")
```

```python
pd.read_csv('employee_latin1.csv')
```

#### as we can seee above there is an error occur due to different enoding 

#### now if we know the  encoding format then we can easily load the file like 


```python
pd.read_csv('employee_latin1.csv',encoding='latin1')
```

#### lets try another if dont know the encoding format 
#### we use `chardet`

```python
# first isntall if chardet is not install in your venv
```

```python
# lets try this code

try:
    df=pd.read_csv('employee_latin1.csv')
except UnicodeDecodeError:
    import chardet

    with open ('employee_latin1.csv','rb') as f:
        enc=chardet.detect(f.read())["encoding"]
    
    df=pd.read_csv('employee_latin1.csv',encoding=enc)
```

```python
df
```

### skip bad lines

##### when ever we got some error while  loading file like `parseError`

```python
pd.read_csv('employee.csv',error_bad_lines=False)
```

##### in modern pandas skip_bad_lines is replaced by `on_bad_line='skip'`

```python
pd.read_csv('employee.csv',on_bad_lines='skip')
```

```python
pd.read_csv('employee_bad.csv')
```

as we see above parse erro occured

```python
pd.read_csv('employee_bad.csv',on_bad_lines='skip')
```

### dtype parameter

##### it is used when without any need in any column the bigger data type is used which causes memory loss hence we useing dtype parameter change the datatype into smalleer one like float to int

```python
pd.read_csv('dataset/salary.csv')
```

```python
pd.read_csv('dataset/salary.csv').info()
```

here we can see that in age and experiance no need of float 

```python
pd.read_csv('dataset/salary.csv',dtype={'age':int,'experience':int}).info()
```

as we can see that dtype is changed now
