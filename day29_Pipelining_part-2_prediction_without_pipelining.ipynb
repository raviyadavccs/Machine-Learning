```python
import pickle   
import numpy as np

```

```python
ohe_sex=pickle.load(open('model/ohe_sex.pkl','rb'))
ohe_embarked=pickle.load(open('model/ohe_embarked.pkl','rb'))
clf=pickle.load(open('model/clf.pkl','rb'))
```

```python
## suppose user input is 
# 	Pclass/Sex/Age/SibSp/Parch/Fare/Embarked

test_input=np.array([2,'male',32.5,4,3,10.5,'S'],dtype=object).reshape(1,7)
```

here we have to specify the dtype because it not mentioned then the numpy array bydefualt change to one singe datatype but our model need both numeric flaot and strign hence `dtype=object` helps to retain datatype


initale the array is one-d vector `(7,)`  but scikit learn model didnt expect in this vector

Example- if 1 passenser have 7 feature then it shold look like (1,7)

```python



# now we have to apply transformation 

test_input_sex=ohe_sex.transform(test_input[:,1].reshape(1,1))
```

above waring is okay it is due to i trianed onehotencoder in  dataframe but now i give numpmy arrray

```python
test_input_embarked=ohe_embarked.transform(test_input[:,6].reshape(1,1)) #  or  test_input[:,-1]

```

```python
test_input_age=test_input[:,2].reshape(1,1)
```

```python
## now transformed all this 
input_remain=test_input[:,[0,3,4,5]]

test_input_transformed=np.concatenate((input_remain,test_input_age,test_input_embarked,test_input_sex),axis=1)
```

```python
test_input_transformed.shape
```

```python
y_prediction=clf.predict(test_input_transformed)
y_prediction
```
