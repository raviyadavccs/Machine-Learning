> ## ***Adding Files to Your Github Repo***



> ### *Step 1: Create the GitHub repository*

Go to GitHub.

Click New Repository.

Repository name:

Machine-Learning

Description (optional):

My Machine Learning learning journey with notes, code, projects, and implementations.

Set it to Public.

Initialize with a README only if your local folder is empty. Otherwise, leave it unchecked.


> ### *Step 2: Open Terminal in your ML folder*

cd path/to/your/ml-folder

For example:

cd D:\Machine Learning

> ### *Step 3: Initialize Git*


git init


> ### *Step 4: Add the GitHub repository*


Replace <username> with your GitHub username.


git remote add origin https://github.com/<username>/Machine-Learning.git


> ### *Step 5: Add all files*

git add .


> ### *Step 6: Commit*

git commit -m "Initial commit"


> ### *Step 7: Push*

If it's your first push:

git branch -M main

git push -u origin main



> ## ***Add new file to Existing Repo***

Step 1: Add your new files

For example, copy or save your new notebook into your repository:


Machine-Learning/

│

├── Lecture-01.ipynb


├── ...

├── Lecture-20.ipynb

├── Lecture-21.ipynb   ← New file


Or if you organize by folders:

Machine-Learning/

└── 07-Regression/

    ├── Lecture-20.ipynb

    └── Lecture-21.ipynb

Step 2: Open a terminal in your repository folder


For example:


cd D:\Machine-Learning

Step 3: Check what changed

Run:


git status


You should see something like:


Untracked files:

    Lecture-21.ipynb

or

Modified:

    README.md

Untracked:

    Lecture-21.ipynb

Step 4: Stage the changes

git add .

Step 5: Create a new commit

git commit -m "Add Lecture 21 - Linear Regression"

Choose a message that describes what you added.

Step 6: Push to GitHub

git push

GitHub will now contain your new commit with Lecture 21.

Your workflow after every lecture

Every time you complete a new lecture:

git status

git add .

git commit -m "Add Lecture XX - Topic Name"

git push

> ## ***Delete any File***

Method 1: Delete the file completely (recommended)

Suppose your repository contains:

Machine-Learning/

├── Lecture-20.ipynb

├── Lecture-21.ipynb

└── test.txt

You want to remove test.txt.

Option A: Delete it using File Explorer or VS Code

Delete test.txt.

Run:

git add .

git commit -m "Remove test.txt"

git push

The file will be removed from your GitHub repository.

Option B: Delete it using Git

git rm test.txt

git commit -m "Remove test.txt"

git push

git rm deletes the file from both your local folder and the Git repository in one step
