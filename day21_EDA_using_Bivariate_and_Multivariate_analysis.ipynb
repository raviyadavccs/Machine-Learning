## 5. HeatMap (Categorical-Categorical)

lets i have to find the number of people survived class wise 

#### we can use `crosstab`  for this 

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

now lets drew this number 

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

as we see the color denotes the number like the lighter color denotes high number whiich can be seen in pclass 3 number of not survided is very high 



#### lets fidn the percent of people in each class survived

```python
titanic.groupby('Pclass').mean()['Survived']
```

above error ocur because this pandas verion try to compute mean of all column but my df contain some string data type column 

sol1.  compute the mean only for the survived column

```python
titanic.groupby('Pclass')['Survived'].mean()
```

as we can see that the  mean of survived is higher in class1 and lowest in class 3 

to see in percentage

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

plot this 

```python
sns.barplot(titanic.groupby('Pclass')['Survived'].mean()*100)
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind="bar")
```

lets fidn the male female survival percentage


```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```
